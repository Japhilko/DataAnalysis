# How to add plots

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
x = np.linspace(0, 3*np.pi, 500)
plt.plot(x, np.sin(x**2))
plt.title('A simple chirp')
plt.show()
# http://stackoverflow.com/questions/19410042/how-to-make-ipython-notebook-inline-matplotlib-graphics
```

# Draw random numbers

```python
mu, sigma = 1,0.1
s = np.random.normal(mu, sigma, 1000)
s2 = np.random.normal(mu, sigma, 1000)
plt.plot(s,s2)
# http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html
```
